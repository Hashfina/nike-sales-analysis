## 1.5 Column Names and Descriptions


Below is a brief description of each column in the Nike sales dataset:

 **Order and product information:**


- Order_ID: Unique identifier for each Nike sales transaction.


- Product_Line: High-level category of the product (for example, Footwear, Apparel, Accessories).


- Product_Name: Specific Nike product sold in the transaction.


- Size: Size of the product (for example, shoe size or clothing size); contains missing values for some records.


**Sales quantity and pricing:**


- Units_Sold: Number of units sold in the transaction.


- MRP: Maximum Retail Price per unit of the product.


- Discount_Applied: Discount amount or percentage applied to the product in that order.


**Revenue, profit, and date:**


- Revenue: Total sales value for the transaction after discount.


- Profit: Profit earned from the transaction after costs.

- Order_Date: Date on which the order was placed; later converted to datetime and used to derive Year and Month.


- Year: Year extracted from Order_Date for time-based analysis.


- Month: Month name extracted from Order_Date for seasonality analysis.


**Customer and channel information:**


- Gender_Category: Customer segment based on gender (for example, Men, Women, Kids).


- Sales_Channel: Channel through which the product was sold (for example, Online, Retail).

- Region: Geographic region where the sale occurred.
