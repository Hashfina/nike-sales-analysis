## 2.1 Data Quality Checks

Missing data in the Nike sales dataset can affect statistical calculations and model performance, especially for key fields like Units_Sold, MRP, Discount_Applied, and Order_Date.

We first need to identify which columns have missing values and how many, so we can plan appropriate handling strategies such as imputation, removal, or business-rule-based fixes.


**Duplicate Rows:**

Duplicate transaction records can distort revenue, profit, and demand analysis, and may indicate data entry or export errors.

Such duplicates should be removed to ensure data integrity before building any insights or models.


The isna().sum() (or isnull().sum()) function is used to count missing values in each column, and duplicated().sum() is used to count duplicate rows.


```python
# 2.2 Data Quality Checks – Missing Values

# Check for missing values in all columns
print("Missing Values Summary:")
missing_values = df.isna().sum()
print(missing_values)
print("=" * 70)

# Columns with missing values
if missing_values.sum() > 0:
    print("\nColumns with missing values:")
    print(missing_values[missing_values > 0])
else:
    print("\nNo missing values found in the dataset.")
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce', dayfirst=True)

df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
df[df['Units_Sold'] < 0].head()
df[df['MRP'] <= 0].head()
df[df['Discount_Applied'] < 0].head()
df[df['Discount_Applied'] > 1].head()   # if discount should be between 0 and 1
df[df['Revenue'] < 0].head()
df[df['Profit'] < 0].head()
```

```python
# remove rows with non‑positive MRP
df = df[df['MRP'] > 0]

# remove rows with impossible discounts (> 1)
df = df[df['Discount_Applied'].between(0, 1)]
```

Cleaning


```python
# 1. Strip spaces and fix casing
cat_cols = ['Gender_Category', 'Product_Line', 'Product_Name', 
            'Size', 'Sales_Channel', 'Region']

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# 2. Optional: Standardize Region (if needed)
df['Region'] = df['Region'].replace({
    'Hyd': 'Hyderabad',
    'Banglore': 'Bengaluru',
    'Bengaluru ': 'Bengaluru',
})

# 3. Optional: Standardize Gender Category
df['Gender_Category'] = df['Gender_Category'].replace({
    'Mens': 'Men',
    'Womens': 'Women',
    'Kids ': 'Kids'
})
```

## Handling invalid and unusual values

First, I checked the numeric columns (`Units_Sold`, `MRP`, `Discount_Applied`, `Revenue`, `Profit`) using `describe()` and filters to find negative or impossible values. I found that some rows had non‑positive MRP values and discount values outside the 0–1 range, which are not realistic for this dataset. 

To improve data quality, I removed rows where `MRP <= 0` and where `Discount_Applied` was less than 0 or greater than 1, because these likely come from data entry errors. I kept negative `Profit` values, since they represent genuine loss‑making transactions that are important for understanding Nike’s performance.


```python
# 1. Convert discount to percentage
df['DiscountPercent'] = df['Discount_Applied'] * 100

# 2. Profit margin (how much profit per 1 unit of revenue)
df['ProfitMargin'] = df['Profit'] / df['Revenue']

# 3. Month–Year for time‑series plots
df['MonthYear'] = df['Order_Date'].dt.to_period('M').astype(str)
```
